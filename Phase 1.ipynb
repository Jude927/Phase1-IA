```python
etudiants = {
    "Junior": [14, 16, 12],
    "Marie": [18, 15, 17]
}

def calculer_moyenne(notes):
    S = 0
    for i in range(len(notes)):
        S += notes[i]
    moy = S / len(notes)
    return moy

   
for nom, notes in etudiants.items():
    moyenne = calculer_moyenne(notes)
    print(f"{nom} : {moyenne}")
```

```python
import matplotlib.pyplot as plt

noms = ["Junior", "Marie", "Paul"]
notes = [14, 18, 12]

plt.bar(noms, notes)
plt.title("Notes des étudiants")
plt.ylabel("Note")
plt.xlabel("Prenoms")
plt.show()
```

```python
mois = ["Jan", "Fev", "Mar", "Avr"]
ventes = [12, 19, 7, 25]

plt.plot(mois, ventes, marker="o")
plt.title("Ventes Triple A")
plt.ylabel("Ventes")
plt.show()
```
